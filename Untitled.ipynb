```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pymongo
import joblib
import os

# Visual settings for your report
sns.set_theme(style="whitegrid")
%matplotlib inline 

# Connect to your MongoDB
client = pymongo.MongoClient("mongodb+srv://admin:1234@cluster0.vbcsfq7.mongodb.net/?appName=Cluster0")
db = client["GlobalCurb"]
logs_col = db["live_training_logs"]
MODEL_PATH = 'adaptive_brain.pkl'

print("Connection Successful! Ready to analyze.")
```

```python
# Fetch all logs
raw_data = list(logs_col.find())

if raw_data:
    df = pd.DataFrame(raw_data)
    # Cleaning: Remove MongoDB ID for a cleaner view
    if '_id' in df.columns:
        df = df.drop(columns=['_id'])
    
    print(f"Total Records Analyzed: {len(df)}")
    display(df.tail()) # Shows the most recent logs
else:
    print("Warning: No logs found.")
```

```python
#Is the AI-generated pricing strategy aligned with user willingness to pay?

plt.figure(figsize=(5, 5))
# Using the 'outcome' column to measure success
colors = {"Accepted": "#2ecc71", "Declined": "#e74c3c"}
sns.countplot(data=df, x='outcome', palette=colors)
plt.title("Price Acceptance Rate (Sentiment Analysis)", fontsize=14)
plt.ylabel("Transaction Count")
plt.show()

# Metric Calculation
acceptance_rate = (len(df[df['outcome']=='Accepted']) / len(df)) * 100
print(f"Current Market Acceptance Rate: {acceptance_rate:.2f}%")
```

```python
#Which urban variable is the most significant driver of parking costs?

if os.path.exists('adaptive_brain.pkl'):
    model = joblib.load('adaptive_brain.pkl')
    # Extracting coefficients from the ensemble
    feat_importances = pd.Series(model.feature_importances_, index=['Lat', 'Lon', 'Hour', 'Quality'])
    
    feat_importances.nlargest(4).plot(kind='barh', color='slategray')
    plt.title("Feature Importance: AI Pricing Weightage")
    plt.xlabel("Relative Importance Score")
    plt.show()
```

```python
#Does the model maintain pricing stability, or is it unpredictable?

plt.figure(figsize=(10, 5))
sns.kdeplot(df['price'], fill=True, color="purple", bw_adjust=0.5)
plt.title("Price Density Estimation (Market Stability)", fontsize=14)
plt.xlabel("Price in ₹")
plt.show()
```

```python
#Is the price truly dynamic? How does it react to time (Hour)?


plt.figure(figsize=(14, 6))

# Aggregate pricing trend (Regression Plot showing scatter + trend line)
sns.regplot(data=df, x='hour', y='price', order=2, scatter_kws={'alpha':0.3, 'color': 'black'}, line_kws={'color':'#e67e22', 'linewidth': 3})

plt.title("Dynamic Pricing Volatility: Regression over 24h Cycle", fontsize=16, fontweight='bold')
plt.xlabel("Hour of the Day (0-23)", fontsize=12)
plt.ylabel("Price (₹)", fontsize=12)
plt.xticks(range(0, 24))
plt.grid(True, which='both', linestyle='--', alpha=0.5)
plt.show()
```
